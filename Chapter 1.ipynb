# Chapter 1: I/O Operations (Image Reading, showing, writing)

Import helpful Libraries

````python
import cv2
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
#print version of Opencv and Python
print('OpenCV-Python Lib Version:', cv2.__version__)
print('Python Version:',sys.version)
````

````python
def displayImage(image,title):
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
    
````

````python
#Reading an image
img_gray = cv2.imread("image.jpg",0) # 0 for grayscal
print(img_gray.shape)
````

````python
#showing an image
displayImage(img_gray,"Gray")
````

````python
img_colored = cv2.imread("image.jpg",1) # for colored Image
print(img_colored.shape)
````

````python
displayImage(img_colored,"Colored Image")
````

### Question 1:
Why Converted BGR to RGB?
### Answer:
Opencv loads image in the format of BGR while the matplotlib library processes it in RGB format.<br>
<b>Note:</b> RGB pixel format is quit standard among MATLAB or other image processing liberaies, but OpenCV uses BGR pixel format.

### Question 2:
Why use using matplotlib to show images instead of standard Opencv imshow function.

### Answer:
As this tutorial is written in jupyter notebook and we needed to show output at the end of our code block, instead Opencv imshow functions pops up another window for display purposes and also crashes down jupyter server, that why we opted for matplotlib library.

<br>

To help with Opencv imshow function here is the code.


```
cv2.imshow('image',img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

1. Use the function cv2.imshow() to display an image in a window. The window automatically fits to the image size.
2. cv2.waitKey() is a keyboard binding function. Its argument is the time in milliseconds. The function waits for specified milliseconds for any keyboard event.
3. cv2.destroyAllWindows() simply destroys all the windows we created. If you want to destroy any specific window, use the function cv2.destroyWindow() where you pass the exact window name as the argument.
